```python
class Vector:
    def __init__(self, elements):  # Vector initialization
        self.elements = elements
    
    def __add__(self, other_vector):  # Vector addition
        return [a + b for a, b in zip(self.elements, other_vector.elements)]
    
    def __sub__(self, other_vector):  # Vector subtraction
        return [a - b for a, b in zip(self.elements, other_vector.elements)]
    
    def __mul__(self, other_vector):  # Vector dot-product operation
        return sum(a * b for a, b in zip(self.elements, other_vector.elements))
```

```python
# Demonstration

A = Vector([3, 1, 4])
B = Vector([5, 6, 7])

print(f"A + B = {A + B}")
print(f"A - B = {A - B}")
print(f"A * B = {A * B}")
```
